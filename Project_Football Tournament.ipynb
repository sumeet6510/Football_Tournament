```python
# 4. Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences.
#    Using the same function, find the team which has maximum and minimum goal difference.
```

```python
# convert columns into list
GoalsFor = list(map(int,df1['GoalsFor']))
GoalsAgainst = list(map(int,df1['GoalsAgainst']))
Team_lsts = list(df1['Team'])
```

```python
# making function
def Goal_diff_count(lsts):
    # create an empty list
    Goal_diff = []
    for i in range(0, len(GoalsFor)):
        Goal_diff.append(GoalsFor[i] - GoalsAgainst[i])
    # map team list with goal_diff    
    dict_teams = dict(zip(Team_lsts, Goal_diff))
    
    return dict_teams
    
```

```python
def keywithmaxval(d):
     """ a) create a list of the dict's keys and values; 
         b) return the key with the max value"""  
     v=list(d.values())
     k=list(d.keys())
     return k[v.index(max(v))]

def keywithminval(d):
     """ a) create a list of the dict's keys and values; 
         b) return the key with the max value"""  
     v=list(d.values())
     k=list(d.keys())
     return k[v.index(min(v))]
```
